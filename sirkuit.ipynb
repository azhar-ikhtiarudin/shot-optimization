```python
from qiskit.circuit import QuantumCircuit, Parameter
```

```python
circ = QuantumCircuit(1,1)
circ.measure(0,0)
circ.draw(style='bw')
```

```python
circ = QuantumCircuit(1,1)
circ.h(0)
circ.measure(0,0)
circ.draw(style='bw')
```

```python
import numpy as np
circ = QuantumCircuit(1,1)
circ.rx(np.pi/2, 0)
circ.measure(0,0)
circ.draw(style='bw')
```

```python
import numpy as np
c = QuantumCircuit(2,2)
c.measure(0,0)
c.measure(1,1)
c.draw(style='bw', reverse_bits=False)
```

```python
import numpy as np
c = QuantumCircuit(2,1)
c.cx(1,0)
c.measure(0,0)
c.draw(style='bw', reverse_bits=False)
```

```python
c = QuantumCircuit(4, 4)
p = Parameter('2t')
c.cx(0,1)
c.cx(1,2)
c.cx(2,3)
c.rz(p, 3)
c.cx(2,3)
c.cx(1,2)
c.cx(0,1)
c.draw(style='bw', reverse_bits=False)
```

```python
c = QuantumCircuit(4, 4)
p = Parameter('2t')
c.h(0)
c.rx(np.pi/2, 1)
c.h(3)

c.cx(0,1)
c.cx(1,2)
c.cx(2,3)
c.rz(p, 3)
c.cx(2,3)
c.cx(1,2)
c.cx(0,1)

c.h(0)
c.rx(np.pi/2, 1)
c.h(3)
c.draw(style='bw', reverse_bits=False)
```
